# RandomForestRegressor

```python
# データ加工・処理・分析ライブラリ
import numpy as np
import numpy.random as random
import pandas as pd

# 可視化ライブラリ
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline

# 機械学習ライブラリ
import sklearn

# 小数第3位まで表示
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

df = pd.read_csv('tforest.csv')
df2 = pd.read_csv('tforest12.csv')

X_train = df.drop(['date'], axis=1)
X_train = X_train.drop(['illumi'], axis=1)
y_train = df['illumi']

X_test = df.drop(['date'], axis=1)
X_test = X_test.drop(['illumi'], axis=1)
y_test = df['illumi']

model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
model.score(X_test, y_test)
```

```python
plt.figure(figsize=(20,10))
plt.plot(y_test,label="True")
plt.plot(predicted, label="predicted")
plt.legend()
```

```python
# 変数の重要度
s = pd.Series(model.feature_importances_, index=X_train.columns)

s.sort_values(ascending=True).plot.barh(color='C0')
```

```python
# 重要度の表
importances = pd.DataFrame(s.sort_values(ascending=False))
importances
```
